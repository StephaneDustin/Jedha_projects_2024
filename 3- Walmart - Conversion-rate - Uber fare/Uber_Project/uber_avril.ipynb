```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Sélection des colonnes latitude et longitude pour le clustering
X = df[['Lat', 'Lon']]


# Calcul du score d'inertie pour différents nombres de clusters
inertia = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(kmeans.inertia_)

# Courbe de la fonction Elbow
plt.figure(figsize=(8, 4))
plt.plot(range(1, 15), inertia, marker='o')
plt.title('Elbow : détermine le nombre optimal de clusters')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Inertie')
plt.xticks(range(1, 15))
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Application de DBSCAN sur df
# Paramètres standards pour commencer: eps=0.01 et min_samples=10
# eps est la distance maximale entre deux échantillons pour qu'ils soient considérés comme dans le même voisinage
# min_samples est le nombre d'échantillons dans un voisinage pour qu'un point soit considéré comme un point central
dbscan = DBSCAN(eps=0.03, min_samples=10, metric="manhattan" ).fit(X)

# Ajout des étiquettes de cluster DBSCAN au DataFrame
df['DBSCAN_Cluster'] = dbscan.labels_

# Affichage du nombre de clusters et du bruit
n_clusters_ = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
n_noise_ = list(dbscan.labels_).count(-1)

print(f'Nombre estimé de clusters: {n_clusters_}')
print(f'Nombre estimé de points de bruit: {n_noise_}')

# Visualisation des résultats de DBSCAN sur une carte avec Plotly
fig_dbscan = px.scatter_mapbox(df, lat='Lat', lon='Lon',
                               color='DBSCAN_Cluster', color_continuous_scale=px.colors.qualitative.D3,
                               title='Clusters identifiés par DBSCAN',
                               mapbox_style='carto-positron')

fig_dbscan.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X, kmeans.labels_)

print(f'Le score moyen de silhouette pour k=7 est : {silhouette_avg}')
```

Essayons Optuna, dont j'ai récemment entendu parler.

```python
import optuna

def objective(trial):
    # Suggérer des valeurs pour les hyperparamètres
    eps = trial.suggest_float('eps', 0.001, 0.1)
    min_samples = trial.suggest_int('min_samples', 2, 20)

    # Appliquer DBScan avec les hyperparamètres suggérés
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan.fit(X)
    
    # Utiliser le score de silhouette comme métrique
    # Note : silhouette_score nécessite au moins 2 labels, exclure le bruit
    unique_labels = set(dbscan.labels_)
    if -1 in unique_labels and len(unique_labels) > 1:
        score = silhouette_score(X, dbscan.labels_)
    else:
        score = -1
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

print(study.best_params)
```

Verdict : le trial numéro 22 est le meilleur combo trouvé par Optuna.
Mais que contient-il comme hyper-paramètres ?? Tadaaaaam !!

```python
#le meilleur essai :
best_trial = study.best_trial

# Informations sur le meilleur essai
print(f'Le meilleur essai est le numéro : {best_trial.number}')
print(f'Les meilleurs hyperparamètres sont : {best_trial.params}')
print(f'Le score de silhouette pour le meilleur essai est : {best_trial.value}')
```

```python
import optuna

def objective(trial):
    # Suggérer des valeurs pour les hyperparamètres
    eps = trial.suggest_float('eps', 0.001, 0.1)
    min_samples = trial.suggest_int('min_samples', 2, 20)

    # Faire tourner DBScan avec le défilement d'hyperparamètres suggérés
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan.fit(X)
    
    # Utiliser le score de silhouette comme métrique
    # Note : silhouette_score nécessite au moins 2 labels, exclure le bruit
    unique_labels = set(dbscan.labels_)
    if -1 in unique_labels and len(unique_labels) > 1:
        score = silhouette_score(X, dbscan.labels_)
    else:
        score = -1
    return score

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

print(study.best_params)
```

```python
dbscan = DBSCAN(eps=0.09906765153638572, min_samples=4, metric="manhattan" ).fit(X)

# Ajout des étiquettes de cluster DBSCAN au DataFrame
df_filtered['DBSCAN_Cluster'] = dbscan.labels_

# Affichage du nombre de clusters et du bruit
n_clusters_ = len(set(dbscan.labels_)) - (1 if -1 in dbscan.labels_ else 0)
n_noise_ = list(dbscan.labels_).count(-1)

print(f'Nombre estimé de clusters: {n_clusters_}')
print(f'Nombre estimé de points de bruit: {n_noise_}')

# Visualisation des résultats de DBSCAN sur une carte avec Plotly
fig_dbscan = px.scatter_mapbox(df_filtered, lat='Lat', lon='Lon',
                               color='DBSCAN_Cluster', color_continuous_scale=px.colors.qualitative.D3,
                               title='Clusters identifiés par DBSCAN',
                               mapbox_style='carto-positron')
airports = {
    'Newark Airport': {'Lat': 40.6895, 'Lon': -74.1745},
    'JFK Airport': {'Lat': 40.6413, 'Lon': -73.7781}
}

import plotly.graph_objects as go

for name, coord in airports.items():
    # Ajouter un marqueur pour l'aéroport
    fig_dbscan.add_trace(go.Scattermapbox(
        lat=[coord['Lat']],
        lon=[coord['Lon']],
        mode='markers',
        marker={'size': 8, 'color': 'red'},
        name=name  # Cette légende apparaîtra dans la légende principale
    ))

    # Ajouter une annotation pour l'aéroport
    fig_dbscan.add_annotation(
        x=coord['Lon'],
        y=coord['Lat'],
        text=name,
        showarrow=False,
        yshift=10
    )

# Ajuster le layout si nécessaire, par exemple pour déplacer la légende
fig_dbscan.update_layout(legend=dict(x=0, y=1))  # Modifier x et y pour positionner la légende

# Calculer le centroïde du cluster unique
cluster_points = df_filtered[df_filtered['DBSCAN_Cluster'] == 0]
centroid = (cluster_points['Lat'].mean(), cluster_points['Lon'].mean())

# Ajouter le centroïde à la figure Plotly existante
fig_dbscan.add_trace(go.Scattermapbox(
    lat=[centroid[0]],
    lon=[centroid[1]],
    mode='markers',
    marker={'size': 15, 'color': 'gold'},
    name='Centroid'
))

# Afficher la figure mise à jour



fig_dbscan.show()
```
